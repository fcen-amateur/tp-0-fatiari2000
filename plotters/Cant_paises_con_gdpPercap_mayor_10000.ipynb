```python
import seaborn.objects as so
from gapminder import gapminder
```

```python
def plot():
    Paises_con_mayor_gdpPercap = gapminder.loc[(gapminder["gdpPercap"] >= 10000) & (gapminder["year"] == 2007)]

    figura = (
        so.Plot(
            data = Paises_con_mayor_gdpPercap, x = "continent"
        )
        .add(so.Bar(), so.Hist(), color = "continent")
        .label(title = "Cantidad de pases con gdpPercap >= 10000 por continente",
               x = "Continente", y = "Cantidad de pases")
)
    
    return dict(descripcion = "El grfico muestra la cantidad de pases con gdpPercap >= 10000 por continente para el ltimo ao del cual hay registro",
               autor = "Fatima Ariana Fernandez",
               figura = figura)
```
